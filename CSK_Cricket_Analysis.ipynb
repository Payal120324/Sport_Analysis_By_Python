```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = "cskdata.csv"
df = pd.read_csv(csv_file_path)

def striker_analysis_graph(striker_name):

    striker_data = df[df['striker'] == striker_name]


    if striker_data.empty:
        print(f"No data available for striker: {striker_name}")
        return


    total_runs = striker_data['runs_of_bat'].sum()


    total_matches = striker_data['match_id'].nunique()


    avg_runs = total_runs / total_matches if total_matches > 0 else 0

    predicted_avg = avg_runs

    avg_runs = round(avg_runs, 2)
    predicted_avg = round(predicted_avg, 2)

    print("Total Matches: " + str(total_matches))
    print("Average Runs:" + str(avg_runs))
    print("Predicted Average:" + str(predicted_avg))


    labels = ['Total Runs', 'Avg Runs/Inning', 'Predicted Avg (Next Year)']
    values = [total_runs, avg_runs, predicted_avg]
    colors = ['blue', 'orange', 'green']

    plt.figure(figsize=(8, 5))
    bars = plt.bar(labels, values, color=colors)


    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 2,
                 f'{bar.get_height()}', ha='center', fontsize=12, fontweight='bold')

    plt.xlabel("Performance Metrics")
    plt.ylabel("Runs")
    plt.title(f"{striker_name} Performance Analysis")
    plt.ylim(0, max(values) + 50)
    plt.show()


striker_name = "Gaikwad"
striker_analysis_graph(striker_name)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


csv_file_path = "cskdata.csv"
df = pd.read_csv(csv_file_path)


match_id_to_no = df[['match_id', 'match_no']].drop_duplicates().set_index('match_id')['match_no']


def get_match_no_by_id(match_id):
    return match_id_to_no.get(match_id, "Match ID not found")


df['total_runs'] = df['runs_of_bat'] + df['extras']


match_summary = df.groupby(['match_id', 'batting_team', 'bowling_team']).agg(
    runs_scored=('total_runs', 'sum')
).reset_index()


csk_matches = match_summary[(match_summary['batting_team'] == 'CSK') | (match_summary['bowling_team'] == 'CSK')]


csk_matches['CSK_scored'] = csk_matches.apply(
    lambda row: row['runs_scored'] if row['batting_team'] == 'CSK' else 0, axis=1
)
csk_matches['CSK_conceded'] = csk_matches.apply(
    lambda ropw: row['runs_scored'] if row['bowling_team'] == 'CSK' else 0, axis=1
)

csk_performance = csk_matches.groupby('match_id').agg(
    CSK_scored=('CSK_scored', 'sum'),
    CSK_conceded=('CSK_conceded', 'sum')
).reset_index()

csk_performance['performance_margin'] = csk_performance['CSK_scored'] - csk_performance['CSK_conceded']

best_match = csk_performance.loc[csk_performance['performance_margin'].idxmax()]
worst_match = csk_performance.loc[csk_performance['performance_margin'].idxmin()]

print("Best Match Performance:")
print(best_match)

print("\nWorst Match Performance:")
print(worst_match)

example_match_id = df['match_id'].iloc[0]
print(f"\nMatch No for Match ID {example_match_id}: {get_match_no_by_id(example_match_id)}")


plt.figure(figsize=(10, 5))
plt.bar(
    csk_performance['match_id'],
    csk_performance['performance_margin'],
    color=['green' if x > 0 else 'red' for x in csk_performance['performance_margin']]
)
plt.xlabel("Match ID")
plt.ylabel("Performance Margin")
plt.title("CSK Match Performance")
plt.axhline(0, color='black', linewidth=1)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'cskdata.csv'
df = pd.read_csv(file_path)


def get_player_performance(striker_name):

    striker_data = df[df['striker'] == striker_name]

    if 'total_runs' not in striker_data.columns:
        striker_data['total_runs'] = striker_data['runs_of_bat'] + striker_data['extras']

    performance = striker_data.groupby('venue').agg({
        'runs_of_bat': 'sum',
        'extras': 'sum',
        'total_runs': 'sum'
    }).reset_index()

    best_venue = performance.loc[performance['runs_of_bat'].idxmax()]

    return performance, best_venue

def plot_player_performance(striker_name):
    performance, best_venue = get_player_performance(striker_name)

    plt.figure(figsize=(12, 6))
    plt.bar(performance['venue'], performance['runs_of_bat'], color='skyblue')
    plt.xlabel('Venue')
    plt.ylabel('Total Runs Scored')
    plt.title(f'Performance of {striker_name} at Each Venue')
    plt.xticks(rotation=45, ha='right')

    plt.axhline(y=best_venue['runs_of_bat'], color='red', linestyle='--', label=f"Best Venue: {best_venue['venue']} ({best_venue['runs_of_bat']} runs)")
    plt.legend()

    plt.tight_layout()
    plt.show()

striker_name = 'Conway'
plot_player_performance(striker_name)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


csv_file_path = "cskdata.csv"
df = pd.read_csv(csv_file_path)


def plot_highest_sixes_and_fours(striker_name):

    required_columns = {'striker', 'runs_of_bat', 'match_id'}
    if not required_columns.issubset(df.columns):
        print(f"Dataset does not have required columns: {required_columns}")
        return


    striker_data = df[df['striker'] == striker_name]

    if striker_data.empty:
        print(f"No data available for the striker: {striker_name}")
        return


    result = striker_data.groupby('match_id').agg(
        Fours=('runs_of_bat', lambda x: (x == 4).sum()),
        Sixes=('runs_of_bat', lambda x: (x == 6).sum())
    ).reset_index()


    total_fours = result['Fours'].sum()
    total_sixes = result['Sixes'].sum()


    labels = ['Fours', 'Sixes']
    values = [total_fours, total_sixes]


    plt.figure(figsize=(8, 6))
    plt.bar(labels, values, color=['blue', 'green'])
    plt.xlabel('Type')
    plt.ylabel('Count')
    plt.title(f'Total Sixes and Fours by {striker_name}')


    for i, v in enumerate(values):
        plt.text(i, v + 1, str(v), ha='center', fontsize=12)

    plt.tight_layout()
    plt.show()


plot_highest_sixes_and_fours("Conway")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('cskdata.csv')


print("Available Columns:", df.columns)


partnership_runs = {}


for match_id, match_data in df.groupby('match_id'):
    batsmen = set()

    for _, row in match_data.iterrows():
        striker = row['striker']
        bowler = row['bowler']


        partnership = tuple(sorted([striker, bowler]))

        if partnership not in partnership_runs:
            partnership_runs[partnership] = 0


        partnership_runs[partnership] += row['runs_of_bat'] + row.get('extras', 0)


partnership_df = pd.DataFrame([
    {'partnership': f"{p[0]} - {p[1]}", 'runs': runs}
    for p, runs in partnership_runs.items()
])

partnership_df = partnership_df.sort_values(by='runs', ascending=False).head(10)


plt.figure(figsize=(12, 6))
plt.barh(partnership_df['partnership'], partnership_df['runs'], color='purple')
plt.xlabel("Runs Scored")
plt.ylabel("Partnerships")
plt.title("Top 10 Partnerships")
plt.gca().invert_yaxis()
plt.show()
```
